```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset

df = pd.read_csv(r'C:\Users\Qazi Saad\OneDrive\Desktop\Internship project\customer_data.csv')
print(df.head())

print(df.info())

# Handle missing values

df['age'] = df['age'].fillna(df['age'].median())
# Check missing values
missing_values = df.isnull().sum()


print("Missing values in each column:")
print(missing_values)

# Check duplicate values
duplicate_values = df.duplicated().sum()
# value counts

gender_values_counts =df['gender'].value_counts()

payment_method_values_counts = df['payment_method'].value_counts()

# summary statistics
summary_statistics = df.describe()

# Data visualization histogram
plt.hist(df['age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Data visualization bar chart 
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

print("Duplicate values:")
print(duplicate_values)
print("Gender values counts:")
print(gender_values_counts)
print("Payment method values counts:")
print(payment_method_values_counts)
print("Descriptive statistics:")
print(summary_statistics)
```
